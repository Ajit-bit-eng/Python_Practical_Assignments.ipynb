# 1. What is hypothesis testing in statistics ?

ANS.

Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It involves formulating a null hypothesis (H₀) and an alternative hypothesis (H₁), collecting data, and using statistical tests to determine which hypothesis is more likely to be true.

# 2. What is the null hypothesis, and how does it differ from the alternative hypothesis ?

ANS.

Null Hypothesis (H₀): Assumes no effect or no difference (e.g., "The new drug has no effect on blood pressure").

Alternative Hypothesis (H₁ or Ha): Represents the claim being tested (e.g., "The new drug reduces blood pressure").

The test determines whether there is enough evidence to reject H₀ in favor of H₁.

# 3. What is the significance level in hypothesis testing, and why is it important ?

ANS.

The significance level (α) is the probability of rejecting the null hypothesis when it is actually true. Common values are 0.05 (5%) or 0.01 (1%). A lower α reduces the chance of a Type 1 error.

Why is it important?

Controls False Positives: It helps limit the chances of mistakenly concluding that there is an effect when there isn't one.

Sets Decision Criteria: It defines the cutoff for determining whether an observed result is due to chance or a real effect.

Common Values: Typical values are 0.05 (5%), 0.01 (1%), or 0.10 (10%), depending on the study’s context.

Balances Risk and Confidence: A lower α (e.g., 0.01) reduces Type I errors but increases the risk of a Type II error (false negative).

# 4. What does a P-value represent in hypothesis testing ?

ANS.

The P-value in hypothesis testing represents the probability of obtaining results at least as extreme as the observed data, assuming that the null hypothesis (H₀) is true. It quantifies the strength of evidence against the null hypothesis.

# 5. How do you interpret the P-value in hypothesis testing ?

ANS.

The P-value helps determine whether the observed data provides enough evidence to reject the null hypothesis (H₀). It represents the probability of getting a result as extreme as the observed one, assuming H₀ is true.

Key Interpretations

P-value ≤ α (e.g., 0.05)

Reject H₀ (null hypothesis).
The result is statistically significant.
There is strong evidence that the alternative hypothesis (H₁) is true.

P-value > α

Fail to reject H₀.
The result is not statistically significant.
There is not enough evidence to support H₁, but it does not prove H₀ is true.

# 6. What are Type 1 and Type 2 errors in hypothesis testing ?

ANS.

1. Type I Error (False Positive)
Occurs when we reject the null hypothesis (H₀) when it is actually true.
This is controlled by the significance level (α), often set at 0.05.
Example: A medical test incorrectly detects a disease in a healthy person.

2. Type II Error (False Negative)
Occurs when we fail to reject the null hypothesis (H₀) when it is actually false.
The probability of a Type II error is β, and 1 - β is the power of the test (ability to detect a true effect).
Example: A medical test fails to detect a disease in a sick person.

# 14. What is a confidence interval, and how is it used to interpret statistical results ?

ANS.

A confidence interval is a range of values, derived from sample data, that is used to estimate an unknown population parameter (like a population mean or proportion). The interval provides a range where we believe the true population parameter lies, with a certain level of confidence.

Estimation of Population Parameters:

Confidence intervals help provide an estimate of a population parameter (e.g., mean, proportion). Instead of stating a single number (point estimate), you provide a range that accounts for sampling variability.

Precision of Estimates:

A narrower confidence interval indicates a more precise estimate of the population parameter, while a wider interval suggests more uncertainty. The width of the interval depends on:

The sample size (larger samples tend to produce narrower intervals).
The variability in the data (more variability results in a wider interval).
The confidence level (higher confidence levels, like 99%, lead to wider intervals).

Hypothesis Testing:

Confidence intervals can also be used for hypothesis testing. For example:

If a null hypothesis value (e.g., a population mean of 50) falls outside the confidence interval, this suggests that the null hypothesis is unlikely to be true at the chosen confidence level (e.g., 95%).

If the null hypothesis value falls within the interval, we do not have sufficient evidence to reject the null hypothesis.

# 19. What is the F-distribution, and when is it used in hypothesis testing ?

ANS.

The F-distribution is a continuous probability distribution that arises frequently in statistical hypothesis testing. It is particularly used when comparing variances across different groups or samples. The distribution is characterized by its positive skewness and the fact that its shape depends on two parameters: degrees of freedom for the numerator and denominator.

The F-distribution is commonly used in the following types of hypothesis tests:

Analysis of Variance (ANOVA):

ANOVA is used to compare the means of three or more groups. It tests the null hypothesis that all group means are equal. The F-statistic is calculated by comparing the variance between the groups (between-group variance) to the variance within the groups (within-group variance). If the F-statistic is large, it indicates that at least one group mean is significantly different from the others.

Example: Testing whether the average scores of students in three different teaching methods are equal.
Testing the Equality of Variances (F-test):

The F-test is used to compare the variances of two samples. The null hypothesis is that the variances of the two populations are equal. The F-statistic is calculated as the ratio of the two sample variances. If the value of the F-statistic is large, it suggests that the variances are significantly different.

Example: Testing whether the variance in heights of two different species of plants is equal.

Regression Analysis:

The F-distribution is used in regression analysis to assess the overall significance of the regression model. The F-test compares the fit of a model with more than one independent variable to a model with no independent variables (i.e., the intercept-only model). If the F-statistic is significantly large, it suggests that at least one of the independent variables has a significant effect on the dependent variable.

Example: Testing whether a set of predictors in a multiple regression model significantly explains the variation in the dependent variable.

General Linear Model:

In complex models like General Linear Models (GLM), the F-distribution is used to test the significance of the model as a whole or to test specific coefficients (groups of coefficients).

# 20. What is an ANOVA test, and what are its assumptions ?

ANS.

The ANOVA (Analysis of Variance) test is a statistical method used to compare means across three or more groups to determine if there are any statistically significant differences between them. ANOVA tests the null hypothesis that all group means are equal. It is an extension of the t-test when comparing more than two groups.

ANOVA relies on several assumptions for the test results to be valid:

Independence:

The samples from each group must be independent of each other. This means the data points in one group should not influence the data points in another group.

Normality:

The data within each group should be approximately normally distributed. This assumption ensures that the variance within the groups is calculated correctly. If the sample size is large enough, the Central Limit Theorem can mitigate deviations from normality.

Homogeneity of Variances (Homogeneity of Variance):

The variance within each group should be approximately equal. This assumption is also called homoscedasticity. If the variances are unequal, the results of ANOVA may not be valid. Levene’s test can be used to check for equality of variances.

# 21. What are the different types of ANOVA tests ?

ANS.

NOVA (Analysis of Variance) has several variations, each suited for different experimental designs and research questions. The main types of ANOVA are:

One-Way ANOVA:

Purpose: Used to test if there are significant differences between the means of three or more independent groups based on one independent variable.

Example: Testing whether three different diets (Diet A, Diet B, Diet C) have different effects on weight loss.

Assumptions: Independence of samples, normality of data within each group, and homogeneity of variances.

Two-Way ANOVA:

Purpose: Used to assess the impact of two independent variables on a dependent variable. It also examines if there is an interaction between the two independent variables.

Example: Testing how both teaching method (Factor 1) and gender (Factor 2) affect students' performance in an exam.
Key Feature: It considers both main effects (individual impact of each factor) and interaction effects (how the factors interact with each other).

Assumptions: Same as One-Way ANOVA, but now the two factors must have independent samples.

Repeated Measures ANOVA:

Purpose: Used when the same subjects are exposed to different conditions or treatments. It compares the means of the same subjects under different conditions over time.

Example: Measuring the performance of athletes before, during, and after a training program.

Assumptions: The observations are dependent, the data is normally distributed for each condition, and sphericity (the variances of the differences between all possible pairs of conditions) is assumed.

Multivariate Analysis of Variance (MANOVA):

Purpose: Used when there are two or more dependent variables, and you want to test if the means of these dependent variables are different across groups based on one or more independent variables.

Example: Testing the effects of diet (independent variable) on both weight loss and blood pressure (dependent variables).

Assumptions: Independence of observations, normality, and equality of covariance matrices for the groups.

Analysis of Covariance (ANCOVA):

Purpose: A blend of ANOVA and regression, ANCOVA is used to examine the effect of one or more independent variables on a dependent variable while controlling for one or more covariates (continuous variables that may influence the dependent variable).

Example: Testing whether different teaching methods (independent variable) affect student performance (dependent variable), while controlling for prior knowledge (covariate).

Assumptions: Homogeneity of regression slopes, normality of residuals, and homogeneity of variance.

Multivariate Analysis of Covariance (MANCOVA):

Purpose: Similar to MANOVA but includes covariates. It examines the influence of independent variables on multiple dependent variables while controlling for covariates.

Example: Testing how different types of diets affect weight loss and cholesterol levels, while controlling for age and gender.

Assumptions: Like MANOVA, but also includes assumptions related to the covariates.

# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

```python
import numpy as np
import scipy.stats as stats

def z_test(sample_data, population_mean, population_std, sample_size):
    # Calculate the sample mean
    sample_mean = np.mean(sample_data)

    # Calculate the Z-statistic
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the p-value for the Z-test
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed test

    return z_stat, p_value

# Sample data and population parameters
sample_data = [110, 115, 120, 125, 130, 135, 140, 145, 150, 155]  # Example sample
population_mean = 120  # Known population mean
population_std = 15    # Known population standard deviation
sample_size = len(sample_data)

# Perform the Z-test
z_stat, p_value = z_test(sample_data, population_mean, population_std, sample_size)

# Print the results
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpret the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

# 2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python

```python
import numpy as np
import scipy.stats as stats

# Simulate random data (e.g., 100 data points with a mean of 50 and standard deviation of 10)
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=50, scale=10, size=100)  # Mean=50, std=10, sample size=100

# Hypothesis Testing (Z-test or T-test)
# Known population parameters (for Z-test), e.g., population mean = 50, population std = 10
population_mean = 50
population_std = 10
sample_size = len(sample_data)

# Function to perform Z-test
def z_test(sample_data, population_mean, population_std, sample_size):
    # Calculate the sample mean
    sample_mean = np.mean(sample_data)

    # Calculate the Z-statistic
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the p-value for the Z-test (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed test

    return z_stat, p_value

# Perform Z-test
z_stat, p_value = z_test(sample_data, population_mean, population_std, sample_size)

# Print the results
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpret the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")

# Alternatively, perform a T-test if population standard deviation is unknown
def t_test(sample_data, population_mean):
    # Calculate the sample mean and sample standard deviation
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation (ddof=1 for sample)

    # Calculate the T-statistic
    t_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(len(sample_data)))

    # Degrees of freedom (sample size - 1)
    df = len(sample_data) - 1

    # Calculate the p-value for the T-test (two-tailed test)
    p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))  # Two-tailed test

    return t_stat, p_value

# Perform T-test
t_stat, t_p_value = t_test(sample_data, population_mean)

# Print the results of the T-test
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value (T-test): {t_p_value:.4f}")

# Interpret the result for T-test
if t_p_value < alpha:
    print("Reject the null hypothesis (T-test): The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis (T-test): The sample mean is not significantly different from the population mean.")
```

# 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean

# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot ?

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Function to perform two-tailed Z-test
def two_tailed_z_test(sample_data, population_mean, population_std, alpha=0.05):
    sample_size = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Calculate Z-statistic
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the two-tailed p-value
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed test

    # Find critical values for a two-tailed test
    z_critical = stats.norm.ppf(1 - alpha / 2)  # Critical Z-score at given alpha

    return z_stat, p_value, z_critical

# Generate a random sample (normal distribution)
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Mean=100, Std=15, n=30

# Define population parameters
population_mean = 105  # Hypothetical population mean
population_std = 15     # Known population standard deviation
alpha = 0.05  # Significance level

# Perform Z-test
z_stat, p_value, z_critical = two_tailed_z_test(sample_data, population_mean, population_std, alpha)

# Print results
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Critical Z-value: ±{z_critical:.2f}")

# Interpret the result
if abs(z_stat) > z_critical:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")

# --- Visualization ---
# Generate a range of Z-values
z_values = np.linspace(-4, 4, 1000)
pdf_values = stats.norm.pdf(z_values)

# Plot the normal distribution
plt.figure(figsize=(10, 5))
plt.plot(z_values, pdf_values, label="Standard Normal Distribution", color='blue')

# Shade the critical regions (left and right tails)
plt.fill_between(z_values, pdf_values, where=(z_values < -z_critical), color='red', alpha=0.5, label="Rejection Region")
plt.fill_between(z_values, pdf_values, where=(z_values > z_critical), color='red', alpha=0.5)

# Plot the Z-statistic
plt.axvline(z_stat, color='green', linestyle='dashed', linewidth=2, label=f"Z-statistic = {z_stat:.2f}")
plt.axvline(-z_critical, color='black', linestyle='dotted', linewidth=2, label=f"Critical Z = ±{z_critical:.2f}")
plt.axvline(z_critical, color='black', linestyle='dotted', linewidth=2)

# Labels and title
plt.title("Two-Tailed Z-Test and Decision Regions")
plt.xlabel("Z-value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()
plt.show()
```

# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing ?

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_type1_type2_errors(population_mean, sample_mean, std_dev, sample_size, alpha=0.05):
    # Compute critical Z-value for given alpha (two-tailed test)
    z_critical = stats.norm.ppf(1 - alpha / 2)

    # Standard error
    std_error = std_dev / np.sqrt(sample_size)

    # Compute rejection region limits
    critical_value_left = population_mean - z_critical * std_error
    critical_value_right = population_mean + z_critical * std_error

    # Generate x-values for plotting
    x_values = np.linspace(population_mean - 4*std_error, population_mean + 4*std_error, 1000)

    # Null and alternative distributions
    null_distribution = stats.norm.pdf(x_values, loc=population_mean, scale=std_error)
    alt_distribution = stats.norm.pdf(x_values, loc=sample_mean, scale=std_error)

    # Plot the distributions
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, null_distribution, label="Null Distribution (H₀)", color='blue')
    plt.plot(x_values, alt_distribution, label="Alternative Distribution (H₁)", color='green')

    # Shade Type 1 Error (alpha - rejection region in null hypothesis)
    plt.fill_between(x_values, null_distribution, where=(x_values <= critical_value_left), color='red', alpha=0.5, label="Type 1 Error (α)")
    plt.fill_between(x_values, null_distribution, where=(x_values >= critical_value_right), color='red', alpha=0.5)

    # Shade Type 2 Error (beta - overlap of null and alternative distributions)
    plt.fill_between(x_values, alt_distribution, where=((x_values > critical_value_left) & (x_values < critical_value_right)), color='blue', alpha=0.5, label="Type 2 Error (β)")

    # Critical value lines
    plt.axvline(critical_value_left, color='black', linestyle='dashed', label=f"Critical Value: {critical_value_left:.2f}")
    plt.axvline(critical_value_right, color='black', linestyle='dashed')

    # Labels and title
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

# Example Usage
population_mean = 100  # Mean under H₀
sample_mean = 105      # Mean under H₁ (assumed shift)
std_dev = 15           # Population standard deviation
sample_size = 30       # Sample size
alpha = 0.05          # Significance level

visualize_type1_type2_errors(population_mean, sample_mean, std_dev, sample_size, alpha)
```

# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
        sample (list or np.array): Sample data.
        confidence (float): Confidence level (default = 0.95 for 95% CI).

    Returns:
        tuple: (Lower bound, Upper bound)
    """
    sample_mean = np.mean(sample)  # Mean of the sample
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (unbiased estimate)
    n = len(sample)  # Sample size

    # Get critical value from t-distribution (used when population std is unknown)
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Calculate margin of error
    margin_of_error = t_critical * (sample_std / np.sqrt(n))

    # Compute confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# --- Example Usage ---
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Simulated data

ci_lower, ci_upper = confidence_interval(sample_data, confidence=0.95)
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
import numpy as np

def bayesian_inference(P_disease, P_positive_given_disease, P_positive_given_no_disease):
    """Computes the posterior probability using Bayes' Theorem."""
    P_no_disease = 1 - P_disease  # Complement of prior probability
    P_positive = (P_positive_given_disease * P_disease) + (P_positive_given_no_disease * P_no_disease)  # Total probability
    P_disease_given_positive = (P_positive_given_disease * P_disease) / P_positive  # Posterior probability
    return P_disease_given_positive

# Define probabilities
P_disease = 0.01  # Prior probability
P_positive_given_disease = 0.95  # Likelihood
P_positive_given_no_disease = 0.05  # False positive rate

# Compute posterior probability
posterior = bayesian_inference(P_disease, P_positive_given_disease, P_positive_given_no_disease)

# Print results
print(f"Probability of having the disease given a positive test result: {posterior:.4f}")

# Explanation of the process:
# 1. Define the prior probability P(A): The probability of having the disease before any test.
# 2. Define the likelihood P(B|A): The probability that the test is positive given the disease.
# 3. Compute the total probability of a positive test P(B): The combined probability of a positive test occurring.
# 4. Compute the posterior probability P(A|B): The updated probability of having the disease given a positive test, using Bayes' Theorem.
```

# 14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution

```python
import numpy as np
import scipy.stats as stats

def goodness_of_fit_test(observed, expected):
    """
    Perform a Chi-Square Goodness-of-Fit test.

    Parameters:
        observed (list or np.array): Observed frequencies.
        expected (list or np.array): Expected frequencies.

    Returns:
        chi2_stat (float): Chi-square test statistic.
        p_value (float): P-value for the test.
    """
    chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)
    return chi2_stat, p_value

# --- Example: Testing if a die is fair ---
# Observed frequencies from rolling a die 120 times
observed_rolls = np.array([18, 22, 20, 15, 25, 20])  # Number of times each face appeared

# Expected frequencies assuming a fair die (each face should appear equally)
total_rolls = np.sum(observed_rolls)
expected_rolls = np.full(6, total_rolls / 6)  # Uniform distribution

# Perform the Chi-Square Goodness-of-Fit Test
chi2_stat, p_value = goodness_of_fit_test(observed_rolls, expected_rolls)

# Print Results
print("Observed Rolls:", observed_rolls)
print("Expected Rolls:", expected_rolls)
print("\nChi-Square Statistic:", chi2_stat)
print("P-Value:", p_value)

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\n Reject the Null Hypothesis: The die is NOT fair.")
else:
    print("\n Fail to Reject the Null Hypothesis: The die appears to be fair.")
```

# 16.  Implement an F-test using Python to compare the variances of two random samples

```python
import numpy as np
import scipy.stats as stats

def f_test(sample1, sample2, alpha=0.05):
    """
    Perform an F-test to compare the variances of two independent samples.

    Parameters:
        sample1 (array-like): First sample data.
        sample2 (array-like): Second sample data.
        alpha (float): Significance level (default is 0.05).

    Returns:
        f_stat (float): Computed F-statistic.
        p_value (float): P-value from the test.
        conclusion (str): Interpretation of the test result.
    """
    # Calculate sample variances
    var1 = np.var(sample1, ddof=1)  # Sample variance (unbiased)
    var2 = np.var(sample2, ddof=1)

    # Compute the F-statistic (larger variance in numerator)
    if var1 > var2:
        f_stat = var1 / var2
        df1, df2 = len(sample1) - 1, len(sample2) - 1
    else:
        f_stat = var2 / var1
        df1, df2 = len(sample2) - 1, len(sample1) - 1

    # Compute the P-value
    p_value = 1 - stats.f.cdf(f_stat, df1, df2)

    # Interpret the result
    if p_value < alpha:
        conclusion = "Reject H₀: The variances are significantly different."
    else:
        conclusion = "Fail to reject H₀: No significant difference in variances."

    return f_stat, p_value, conclusion

# --- Simulate Random Data for Testing ---
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10, n=30
sample2 = np.random.normal(loc=50, scale=15, size=30)  # Mean=50, Std=15, n=30

# Perform the F-test
f_stat, p_value, result = f_test(sample1, sample2)

# Print Results
print(f"Sample 1 Variance: {np.var(sample1, ddof=1):.2f}")
print(f"Sample 2 Variance: {np.var(sample2, ddof=1):.2f}")
print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4f}")
print("\nResult:", result)
```

# 17. Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results

```python
import numpy as np
import scipy.stats as stats

def anova_test(*groups, alpha=0.05):
    """
    Perform a One-Way ANOVA test to compare means across multiple groups.

    Parameters:
        *groups (arrays): Data for each group.
        alpha (float): Significance level (default is 0.05).

    Returns:
        f_stat (float): ANOVA F-statistic.
        p_value (float): P-value from the test.
        conclusion (str): Interpretation of the test result.
    """
    f_stat, p_value = stats.f_oneway(*groups)

    # Interpretation
    if p_value < alpha:
        conclusion = "Reject H₀: At least one group mean is significantly different."
    else:
        conclusion = "Fail to reject H₀: No significant difference between group means."

    return f_stat, p_value, conclusion

# --- Simulate Random Data for 3 Groups ---
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10, n=30
group2 = np.random.normal(loc=55, scale=10, size=30)  # Mean=55, Std=10, n=30
group3 = np.random.normal(loc=60, scale=10, size=30)  # Mean=60, Std=10, n=30

# Perform the ANOVA test
f_stat, p_value, result = anova_test(group1, group2, group3)

# Print Results
print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4f}")
print("\nResult:", result)
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def anova_test_plot(*groups, alpha=0.05):
    """
    Perform a One-Way ANOVA test and visualize the group distributions.

    Parameters:
        *groups (arrays): Data for each group.
        alpha (float): Significance level (default is 0.05).

    Returns:
        f_stat (float): ANOVA F-statistic.
        p_value (float): P-value from the test.
        conclusion (str): Interpretation of the test result.
    """
    f_stat, p_value = stats.f_oneway(*groups)

    # Interpretation
    if p_value < alpha:
        conclusion = "🔴 Reject H₀: At least one group mean is significantly different."
    else:
        conclusion = "✅ Fail to reject H₀: No significant difference between group means."

    # Create DataFrame for Visualization
    data = []
    for i, group in enumerate(groups):
        data.extend([(f'Group {i+1}', value) for value in group])

    df = pd.DataFrame(data, columns=["Group", "Value"])

    # Plot the distributions using a boxplot
    plt.figure(figsize=(8, 6))
    sns.boxplot(x="Group", y="Value", data=df, palette="Set2")
    plt.title("One-Way ANOVA: Group Comparisons")
    plt.xlabel("Groups")
    plt.ylabel("Values")
    plt.grid(True)
    plt.show()

    return f_stat, p_value, conclusion

# --- Simulate Random Data for 3 Groups ---
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10, n=30
group2 = np.random.normal(loc=55, scale=10, size=30)  # Mean=55, Std=10, n=30
group3 = np.random.normal(loc=60, scale=10, size=30)  # Mean=60, Std=10, n=30

# Perform the ANOVA test and plot the results
f_stat, p_value, result = anova_test_plot(group1, group2, group3)

# Print Results
print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4f}")
print("\nResult:", result)
```

# 19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

```python
import numpy as np
import scipy.stats as stats

def check_anova_assumptions(*groups, alpha=0.05):
    """
    Check ANOVA assumptions: normality and equal variance.

    Parameters:
        *groups (arrays): Data for each group.
        alpha (float): Significance level (default is 0.05).

    Returns:
        dict: Results of the assumption tests.
    """
    results = {}

    # --- 1. Normality Check (Shapiro-Wilk Test) ---
    normality_p_values = [stats.shapiro(group)[1] for group in groups]
    results['normality'] = all(p > alpha for p in normality_p_values)

    # --- 2. Equal Variance Check (Levene’s Test) ---
    levene_stat, levene_p = stats.levene(*groups)
    results['equal_variance'] = levene_p > alpha

    return results

# --- Simulate Random Data ---
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)
group3 = np.random.normal(loc=60, scale=10, size=30)

# Check ANOVA assumptions
assumptions = check_anova_assumptions(group1, group2, group3)

# Print Results
print("ANOVA Assumption Checks:")
print(f"Normality Passed: {assumptions['normality']}")
print(f"Equal Variance Passed: {assumptions['equal_variance']}")
```

# 20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

# --- Simulate Random Data ---
np.random.seed(42)

# Factors: Diet Type (Keto, Vegan, Normal) & Exercise Level (Low, Moderate, High)
diet_types = ['Keto', 'Vegan', 'Normal']
exercise_levels = ['Low', 'Moderate', 'High']

# Create a dataset
data = []
for diet in diet_types:
    for exercise in exercise_levels:
        weight_loss = np.random.normal(loc=5, scale=2, size=10)  # Simulating weight loss
        for loss in weight_loss:
            data.append([diet, exercise, loss])

# Convert to DataFrame
df = pd.DataFrame(data, columns=['Diet', 'Exercise', 'Weight_Loss'])

# --- Perform Two-Way ANOVA ---
model = ols('Weight_Loss ~ C(Diet) + C(Exercise) + C(Diet):C(Exercise)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Print Results
print("\n🔹 Two-Way ANOVA Results:")
print(anova_table)

# --- Visualization: Boxplot for Each Factor ---
plt.figure(figsize=(10, 6))
sns.boxplot(x="Diet", y="Weight_Loss", hue="Exercise", data=df, palette="Set2")
plt.title("Two-Way ANOVA: Effect of Diet & Exercise on Weight Loss")
plt.xlabel("Diet Type")
plt.ylabel("Weight Loss (kg)")
plt.legend(title="Exercise Level")
plt.grid(True)
plt.show()
```

# 21. Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def plot_f_distribution(df1, df2, alpha=0.05):
    """
    Visualizes the F-distribution and highlights the critical region.

    Parameters:
        df1 (int): Degrees of freedom for the numerator.
        df2 (int): Degrees of freedom for the denominator.
        alpha (float): Significance level (default 0.05).
    """
    x = np.linspace(0, 5, 1000)  # X-axis values
    y = stats.f.pdf(x, df1, df2)  # F-distribution PDF

    # Plot F-distribution
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=f"F-Distribution (df1={df1}, df2={df2})", color="blue")

    # Critical F-value at 95% confidence level
    f_critical = stats.f.ppf(1 - alpha, df1, df2)

    # Highlight the critical region
    x_critical = np.linspace(f_critical, 5, 500)
    y_critical = stats.f.pdf(x_critical, df1, df2)
    plt.fill_between(x_critical, y_critical, color='red', alpha=0.5, label=f"Critical Region (α={alpha})")

    # Labels and legend
    plt.axvline(f_critical, color="red", linestyle="dashed", label=f"Critical F={f_critical:.2f}")
    plt.title("F-Distribution and Critical Region")
    plt.xlabel("F-value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example: Degrees of Freedom (df1=5, df2=10)
plot_f_distribution(df1=5, df2=10)
```

# 22. Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# --- Simulate Random Data for 3 groups (e.g., Diet A, Diet B, Diet C) ---
np.random.seed(42)
group_a = np.random.normal(loc=30, scale=5, size=30)
group_b = np.random.normal(loc=35, scale=5, size=30)
group_c = np.random.normal(loc=40, scale=5, size=30)

# Create a DataFrame for the data
data = pd.DataFrame({
    'Diet': ['Diet A']*30 + ['Diet B']*30 + ['Diet C']*30,
    'Weight_Loss': np.concatenate([group_a, group_b, group_c])
})

# --- Perform One-Way ANOVA ---
f_statistic, p_value = stats.f_oneway(group_a, group_b, group_c)

# Print the ANOVA results
print("🔹 One-Way ANOVA Results:")
print(f"F-statistic: {f_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# --- Interpretation ---
if p_value < 0.05:
    print("Reject the null hypothesis: At least one group mean is different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between group means.")

# --- Visualization with Boxplot ---
plt.figure(figsize=(8, 6))
sns.boxplot(x='Diet', y='Weight_Loss', data=data, palette='Set3')
plt.title("One-Way ANOVA: Comparison of Weight Loss Across Diets")
plt.xlabel("Diet Type")
plt.ylabel("Weight Loss (kg)")
plt.grid(True)
plt.show()
```

# 23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means

```python
import numpy as np
import scipy.stats as stats

# --- Simulate Random Data from a Normal Distribution ---
np.random.seed(42)
population_mean = 50  # Assume the population mean is 50
population_std = 10   # Standard deviation of the population
sample_size = 30      # Sample size
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# --- Perform One-Sample t-Test ---
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
t_statistic, p_value = stats.ttest_1samp(sample_data, population_mean)

# --- Print Results ---
print("🔹 One-Sample t-Test Results:")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"t-Statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# --- Interpretation ---
if p_value < 0.05:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

# 24.  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

```python
import numpy as np
import scipy.stats as stats

# --- Simulate Random Data (sample data from normal distribution) ---
np.random.seed(42)
population_variance = 25   # Hypothetical population variance (σ²₀)
sample_size = 30           # Sample size
sample_data = np.random.normal(loc=50, scale=np.sqrt(population_variance), size=sample_size)

# --- Calculate Sample Variance ---
sample_variance = np.var(sample_data, ddof=1)

# --- Perform Chi-Square Test for Population Variance ---
chi_square_statistic = (sample_size - 1) * sample_variance / population_variance

# --- Determine the critical value for Chi-Square test ---
alpha = 0.05  # Significance level
df = sample_size - 1  # Degrees of freedom
chi_square_critical_upper = stats.chi2.ppf(1 - alpha / 2, df)
chi_square_critical_lower = stats.chi2.ppf(alpha / 2, df)

# --- Print Results ---
print("🔹 Chi-Square Test for Population Variance:")
print(f"Sample Variance: {sample_variance:.2f}")
print(f"Chi-Square Statistic: {chi_square_statistic:.2f}")
print(f"Critical Chi-Square Values: Lower = {chi_square_critical_lower:.2f}, Upper = {chi_square_critical_upper:.2f}")

# --- Interpretation ---
if chi_square_statistic < chi_square_critical_lower or chi_square_statistic > chi_square_critical_upper:
    print("Reject the null hypothesis: The population variance is different from the specified value.")
else:
    print("Fail to reject the null hypothesis: The population variance is not significantly different from the specified value.")
```

# 25.  Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

```python
import numpy as np
import scipy.stats as stats

# --- Simulate Data for Two Groups ---
# Sample sizes for the two groups
n1 = 1000  # Group 1 sample size
n2 = 1000  # Group 2 sample size

# Number of successes in each group
successes_group1 = 520  # Successes in group 1
successes_group2 = 480  # Successes in group 2

# Sample proportions
p1 = successes_group1 / n1  # Proportion of success in group 1
p2 = successes_group2 / n2  # Proportion of success in group 2

# Combined proportion (P)
P = (successes_group1 + successes_group2) / (n1 + n2)

# --- Z-Test for Proportions ---
# Z-test statistic formula
z_statistic = (p1 - p2) / np.sqrt(P * (1 - P) * (1 / n1 + 1 / n2))

# --- Calculate P-value ---
# Two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))  # Double the area in one tail

# --- Print Results ---
print("🔹 Z-Test for Proportions Results:")
print(f"Proportion of successes in group 1: {p1:.4f}")
print(f"Proportion of successes in group 2: {p2:.4f}")
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# --- Interpretation ---
if p_value < 0.05:
    print("Reject the null hypothesis: The proportions of success in the two groups are significantly different.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the proportions of success in the two groups.")
```

# 26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# --- Simulate Two Datasets with Different Variances ---
np.random.seed(42)

# Sample sizes
n1 = 100  # Group 1 size
n2 = 100  # Group 2 size

# Population standard deviations
std1 = 10  # Standard deviation for group 1
std2 = 15  # Standard deviation for group 2

# Simulating random data from normal distributions
data_group1 = np.random.normal(loc=50, scale=std1, size=n1)
data_group2 = np.random.normal(loc=50, scale=std2, size=n2)

# --- Calculate Sample Variances ---
var1 = np.var(data_group1, ddof=1)  # Sample variance for group 1
var2 = np.var(data_group2, ddof=1)  # Sample variance for group 2

# --- Perform F-Test ---
F_statistic = var1 / var2  # F-statistic is the ratio of sample variances
df1 = n1 - 1  # Degrees of freedom for group 1
df2 = n2 - 1  # Degrees of freedom for group 2

# --- Calculate P-value ---
# Right-tailed test: P-value is the probability of obtaining a test statistic as extreme as observed
p_value = 1 - stats.f.cdf(F_statistic, df1, df2)

# --- Print Results ---
print("🔹 F-Test for Comparing Variances Results:")
print(f"Sample Variance for Group 1: {var1:.2f}")
print(f"Sample Variance for Group 2: {var2:.2f}")
print(f"F-Statistic: {F_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# --- Interpretation ---
if p_value < 0.05:
    print("Reject the null hypothesis: The variances of the two groups are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances of the two groups are not significantly different.")

# --- Visualization of the F-distribution ---
x = np.linspace(0, 5, 1000)
y = stats.f.pdf(x, df1, df2)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'F-distribution (df1={df1}, df2={df2})')
plt.axvline(x=F_statistic, color='r', linestyle='--', label=f'F-statistic: {F_statistic:.2f}')
plt.fill_between(x, y, where=(x >= F_statistic), color='r', alpha=0.3, label=f'Rejection region (p-value = {p_value:.4f})')

plt.title("F-Test for Comparing Variances")
plt.xlabel("F-statistic")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

# 27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# --- Simulate Observed Data ---
# Suppose we are testing if the data follows a uniform distribution
np.random.seed(42)

# Simulate 1000 observations in 4 categories (e.g., 4 outcomes in a dice roll)
observed_data = np.random.choice([1, 2, 3, 4], size=1000)

# Calculate observed frequencies
observed_frequencies = [np.sum(observed_data == i) for i in [1, 2, 3, 4]]

# --- Define Expected Frequencies ---
# Under the null hypothesis, each category has an equal probability (uniform distribution)
expected_frequencies = [1000 / 4] * 4  # Expected frequencies for a uniform distribution

# --- Perform Chi-Square Goodness of Fit Test ---
chi2_statistic, p_value = stats.chisquare(observed_frequencies, expected_frequencies)

# --- Print Results ---
print("🔹 Chi-Square Goodness of Fit Test Results:")
print(f"Observed Frequencies: {observed_frequencies}")
print(f"Expected Frequencies: {expected_frequencies}")
print(f"Chi-Square Statistic: {chi2_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# --- Interpretation ---
if p_value < 0.05:
    print("Reject the null hypothesis: The observed data does not follow the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data follows the expected distribution.")

# --- Visualization of the Observed vs Expected Frequencies ---
categories = [1, 2, 3, 4]
width = 0.3  # Bar width

plt.figure(figsize=(8, 6))
plt.bar(categories, observed_frequencies, width=width, label='Observed', color='blue', alpha=0.6, align='center')
plt.bar(np.array(categories) + width, expected_frequencies, width=width, label='Expected', color='red', alpha=0.6, align='center')

plt.title("Chi-Square Goodness of Fit Test")
plt.xlabel("Categories")
plt.ylabel("Frequencies")
plt.legend()
plt.xticks(categories)
plt.grid(True)
plt.show()
```
